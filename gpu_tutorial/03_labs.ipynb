## Notebook Sources
- Will be using [google colab](https://colab.research.google.com) as our platform.
- Clicking on the link to each notebook should open up an empty (or the solutions) notebooks in colab.
- Largely based on the tutorial [here](https://github.com/NVIDIA/accelerated-computing-hub/blob/main/tutorials/accelerated-python/notebooks/syllabi/cuda_python__cupy_cudf_cccl_kernels__8_hours.ipynb)

### Hints
- Make sure to select "T4 GPU" as the platform so that we can use a GPU for testing.  You should see a RAM and Disk graph in the upper right corner.
- You may need to "trust" the notebook as not being authored from google.  Click on "Run anyway" if this comes up.
- If it complains that you have too many sessions open, click on manage sessions and click on "Terminate other sessions"
